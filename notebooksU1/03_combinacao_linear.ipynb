## Introdução

Uma das ferramentas mais poderosas na Álgebra Linear é a criação de novos vetores a partir da mescla de vetores existentes. Esse processo é chamado de **Combinação Linear**. Mais do que uma simples definição abstrata, a combinação linear é o "elo perdido" que conecta espaços vetoriais com a resolução de **Sistemas Lineares**.

Ao final deste tutorial, você será capaz de:
1. Definir formalmente o conceito de combinação linear.
2. Escrever um vetor em função de outros.
3. Compreender a ligação entre combinações lineares e matrizes.
4. Resolver computacionalmente coeficientes de combinações com NumPy.


## 1. O que é uma Combinação Linear?

Dado um conjunto de vetores $\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k\}$ em um espaço vetorial $V$, dizemos que um vetor $\mathbf{u}$ é uma **combinação linear** desse conjunto se é possível expressar $\mathbf{u}$ na forma:

$$\mathbf{u} = c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \ldots + c_k\mathbf{v}_k$$

Onde $c_1, c_2, \ldots, c_k$ são escalares (números reais).

### A ligação com Sistemas Lineares
Se você tentar descobrir quais são os escalares $c_i$ necessários para formar um certo vetor $\mathbf{u}$, você estará essencialmente **resolvendo um sistema linear**.

Se os vetores formam as colunas de uma matriz $A$, teremos que:
$$A\mathbf{c} = \mathbf{u}$$

Neste cenário:
- A matriz $A$ é formada pelos vetores da combinação colocados em colunas.
- $\mathbf{c}$ é o vetor formado pelos escalares incertos.
- $\mathbf{u}$ é o resultado da combinação (termos independentes).


## 2. Abordagem Computacional com NumPy

Vamos verificar no Python se um vetor $\mathbf{u} = (7, 11)$ pode ser escrito como combinação linear de $\mathbf{v}_1 = (1, 2)$ e $\mathbf{v}_2 = (2, 3)$. 

Para isso, tentamos resolver:
$$c_1\begin{pmatrix}1 \\ 2\end{pmatrix} + c_2\begin{pmatrix}2 \\ 3\end{pmatrix} = \begin{pmatrix}7 \\ 11\end{pmatrix}$$


```python
import numpy as np

# A matriz cujas colunas são v1 e v2
A = np.array([
    [1, 2],
    [2, 3]
])

# O vetor resultante u
u = np.array([7, 11])

# Resolvendo o sistema A*c = u usando numpy.linalg
try:
    c = np.linalg.solve(A, u)
    print("O vetor u PODE ser escrito como combinação linear!")
    print(f"Os escalares são: c1 = {c[0]:.2f}, c2 = {c[1]:.2f}")
    
    # Conferindo a volta:
    v1 = np.array([1, 2])
    v2 = np.array([2, 3])
    print(f"Verificação: {c[0]}*v1 + {c[1]}*v2 = {c[0]*v1 + c[1]*v2}")
except np.linalg.LinAlgError:
    print("O sistema não tem solução. O vetor u NÃO É combinação linear de v1 e v2.")
```

## 3. Quando um vetor NÃO é combinação linear?

É possível que alguns vetores nunca consigam chegar no vetor final, não importa como você os estique e some. Nesse caso, dizemos que o sistema linear é impossível (SI).


```python
# Exemplo de falha de combinação (vetores paralelos)
v1_paralelo = np.array([1, 2])
v2_paralelo = np.array([2, 4])  

A_falha = np.column_stack((v1_paralelo, v2_paralelo))
u_falha = np.array([5, 5])

try:
    c_falha = np.linalg.solve(A_falha, u_falha)
except np.linalg.LinAlgError as e:
    print("Erro capturado:", e)
    print("Conclusão: (5,5) NÃO pode ser formado pela soma das colunas.")
```
