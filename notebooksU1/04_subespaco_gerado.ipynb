No tutorial anterior vimos como combinar vetores. E se decidíssemos colecionar **TODOS** os vetores possíveis que podem ser formados por uma combinação linear? Essa "coleção monstra" é o que chamamos de **Subespaço Gerado** ou **Span**.

## 1. O que é Conjunto Gerador (Span)?

Dado um conjunto de vetores $S = \{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k\} \subseteq V$, o **Subespaço Gerado** por $S$, denotado por $\text{span}(S)$ ou $\langle S \rangle$, é o conjunto de todas as combinações lineares possíveis compostas por esses vetores.

Formalmente:
$$\text{span}(\{\mathbf{v}_1, \ldots, \mathbf{v}_k\}) = \{ c_1\mathbf{v}_1 + \ldots + c_k\mathbf{v}_k \mid c_1, \ldots, c_k \in \mathbb{R} \}$$

### Interpretando Geometricamente na Prática:
- Span de um vetor nulo: É apenas a origem.
- Span de 1 vetor (não nulo): É uma reta infinita que passa na origem.
- Span de 2 vetores que não "apontam" para o mesmo lugar: É um plano infinito inteiro que recorta a origem.
- Span de 3 vetores "independentes" no $\mathbb{R}^3$: Recobre o $\mathbb{R}^3$ todinho!


## 2. O Espaço Coluna e o Espaço Linha de uma Matriz

Quando estruturamos as coisas como uma Matriz, chegamos a nomes especiais para os Span's formados:
- **Espaço Coluna ($Col\ A$):** É o span gerado pelas colunas da matriz $A$. Se $A\mathbf{x} = \mathbf{b}$ tem solução, significa que $\mathbf{b}$ pertence ao espaço coluna de $A$.
- **Espaço Linha ($Row\ A$):** É o span formado pelas linhas da matriz $A$.

Usaremos o módulo SymPy do Python, que possui ferramentas nativas maravilhosas para nos dizer quais colunas "importam" e de fato formam o span.


```python
import sympy as sp

# Suponha uma matriz cujas colunas formam um conjunto S
A = sp.Matrix([
    [1, 2, 3],
    [2, 4, 6],
    [-1, 1, 0]
])

print("Matriz A:")
sp.pprint(A)

# O espaço coluna é gerado pelas colunas de A.
# O método columnspace() do SymPy extrai quais colunas formam de verdade o Span
col_space = A.columnspace()

print("\nVetores base que atuam como Geradores Principais do Span (Espaço Coluna):")
for col in col_space:
    sp.pprint(col)
    
print("\nNote que a 3º coluna era apenas a soma da primeira com a segunda!")
```

## 3. Visualização: Span de Dois Vetores
Se colocarmos vetores $\mathbf{u}$ e $\mathbf{v}$ (não paralelos) no $\mathbb{R}^3$, o Span deles cobrirá geometricamente todo um plano. Como plotar um Span infinito em Python? Plotamos uma grade ampla de combinações aleatórias.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

sns.set_theme(style="white", context="talk")

# Dois vetores no plano z=0
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Gerando múltiplas combinações lineares (simulando a formação do Span)
cs_u = np.linspace(-5, 5, 20)
cs_v = np.linspace(-5, 5, 20)

pontos_span = []
for c1 in cs_u:
    for c2 in cs_v:
        pontos_span.append(c1*u + c2*v)

pontos_span = np.array(pontos_span)

# Plotando os pontos que o Span forma (Resultará no plano cartesiano XY)
ax.scatter(pontos_span[:,0], pontos_span[:,1], pontos_span[:,2], color='skyblue', alpha=0.3, s=5)
ax.quiver(0, 0, 0, *u, color='b', linewidth=3, label='u')
ax.quiver(0, 0, 0, *v, color='g', linewidth=3, label='v')

ax.set_title("O plano XY é o Span dos vetores (1,0,0) e (0,1,0)")
ax.set_zlim([-2, 2])
plt.legend()
plt.show()
```
