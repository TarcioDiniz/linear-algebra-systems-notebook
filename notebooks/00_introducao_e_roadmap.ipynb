# 4. Exemplo Motivador

Para começar, veja como o Python resolve instantaneamente um problema que levaria tempo no papel.

**O Desafio:** "A soma de dois números é 10 e a diferença entre eles é 2. Quais são esses números?"
Em matemática, temos:
$$
\begin{cases}
x + y = 10 \\
x - y = 2
\end{cases}
$$

Observe o código abaixo para ver a solução numérica e visual:


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sp

# Configuração do Seaborn para gráficos mais bonitos
sns.set_theme(style="whitegrid", palette="muted")

# 1. Definindo as matrizes do sistema $A\mathbf{x} = \mathbf{b}$
A = np.array([[1, 1], 
              [1, -1]], dtype=float)
b = np.array([10, 2], dtype=float)

# 2. Resolvendo o sistema
x_sol = np.linalg.solve(A, b)
display(sp.Matrix(x_sol))

# 3. Visualização com Seaborn
x_vals = np.linspace(0, 10, 100)
y1 = 10 - x_vals
y2 = x_vals - 2

plt.figure(figsize=(8, 5))
sns.lineplot(x=x_vals, y=y1, label='$x + y = 10$', linewidth=2)
sns.lineplot(x=x_vals, y=y2, label='$x - y = 2$', linewidth=2)

# Destacando a solução
plt.scatter(x_sol[0], x_sol[1], color='red', s=100, zorder=5, edgecolor='white', label=f'Interseção {x_sol}')

# Ajustes estéticos finais
plt.axhline(0, color='black', lw=0.8, alpha=0.5)
plt.axvline(0, color='black', lw=0.8, alpha=0.5)
plt.grid(True, linestyle='--')
plt.title("Visualização do Sistema 2x2", fontsize=14, pad=15)
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$y$", fontsize=12)
plt.legend(frameon=True, shadow=True)
sns.despine()
plt.tight_layout()
plt.show()
```
