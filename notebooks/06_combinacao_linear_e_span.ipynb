## Introdução

Neste tutorial, exploraremos a operação fundamental de **Combinação Linear** e o conceito de **Subespaço Gerado (Span)**. A combinação linear é o mecanismo que une vetores individuais para formar novos elementos de um espaço vetorial, estabelecendo o elo entre vetores e a resolução de sistemas lineares.

Ao final deste tutorial, o estudante será capaz de:
1. Definir formalmente a combinação linear de vetores.
2. Identificar se um vetor pertence ao subespaço gerado por um conjunto dado.
3. Compreender a relação entre o Span de colunas e a consistência de sistemas lineares.
4. Resolver computacionalmente os coeficientes de combinações lineares utilizando NumPy e SymPy.

**Referência principal:** BOLDRINI, J. L. et al. *Álgebra Linear*. 3ª ed. São Paulo: Harper & Row do Brasil, 1980.

# 1. Definição de Combinação Linear

Dado um conjunto de vetores $\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k\}$ em um espaço vetorial $V$, dizemos que um vetor $\mathbf{u}$ é uma **combinação linear** desse conjunto se é possível expressar $\mathbf{u}$ na forma (BOLDRINI et al., 1980):

$$\mathbf{u} = c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \ldots + c_k\mathbf{v}_k$$

Onde $c_1, c_2, \ldots, c_k$ são escalares (números reais).

### A ligação com Sistemas Lineares
Se você tentar descobrir quais são os escalares $c_i$ necessários para formar um certo vetor $\mathbf{u}$, você estará essencialmente **resolvendo um sistema linear**.

Se os vetores formam as colunas de uma matriz $A$, teremos que (BOLDRINI et al., 1980):
$$Ac = u$$

Neste cenário:
- A matriz $A$ é formada pelos vetores da combinação colocados em colunas.
- $c$ é o vetor formado pelos escalares incertos.
- $u$ é o resultado da combinação (termos independentes).


# 2. Abordagem Computacional com NumPy

Vamos verificar no Python se um vetor $\mathbf{u} = (7, 11)$ pode ser escrito como combinação linear de $\mathbf{v}_1 = (1, 2)$ e $\mathbf{v}_2 = (2, 3)$. 

Para isso, tentamos resolver:
$$c_1\begin{pmatrix}1 \\ 2\end{pmatrix} + c_2\begin{pmatrix}2 \\ 3\end{pmatrix} = \begin{pmatrix}7 \\ 11\end{pmatrix}$$


```python
import numpy as np
from sympy import Matrix

# A matriz cujas colunas são v1 e v2
A = np.array([
    [1, 2],
    [2, 3]
])

# O vetor resultante u
u = np.array([7, 11])

# Resolvendo o sistema A*c = u usando numpy.linalg
try:
    c = np.linalg.solve(A, u)
    print("O vetor u PODE ser escrito como combinação linear!")
    print(f"Os escalares são: c1 = {c[0]:.2f}, c2 = {c[1]:.2f}")
    
    # Conferindo a volta:
    v1 = np.array([1, 2])
    v2 = np.array([2, 3])
    display(Matrix(c[0]*v1 + c[1]*v2))
except np.linalg.LinAlgError:
    print("O sistema não tem solução. O vetor u NÃO É combinação linear de v1 e v2.")
```

# 3. Quando um vetor NÃO é combinação linear?

É possível que alguns vetores nunca consigam chegar no vetor final, não importa como você os escalone e some. Nesse caso, dizemos que o sistema linear é impossível (SI) (BOLDRINI et al., 1980).


```python
# Exemplo de falha de combinação (vetores paralelos)
v1_paralelo = np.array([1, 2])
v2_paralelo = np.array([2, 4])  

A_falha = np.column_stack((v1_paralelo, v2_paralelo))
u_falha = np.array([5, 5])

try:
    c_falha = np.linalg.solve(A_falha, u_falha)
except np.linalg.LinAlgError as e:
    print("Erro capturado:", e)
    print("Conclusão: (5,5) NÃO pode ser formado pela soma das colunas.")
```
