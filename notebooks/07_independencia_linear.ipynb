## Introdução

Neste tutorial, abordaremos os conceitos de **Independência Linear (LI)** e **Dependência Linear (LD)**. Estes conceitos permitem identificar a existência de vetores redundantes em um conjunto e são fundamentais para a construção de bases em espaços vetoriais.

Ao final deste tutorial, o estudante será capaz de:
1. Definir formalmente a independência linear de um conjunto de vetores.
2. Identificar vetores redundantes em uma matriz utilizando o escalonamento (RREF).
3. Aplicar o teste do determinante para verificação de independência linear.
4. Conectar o conceito de LI com a solução de sistemas homogêneos.

**Referência principal:** BOLDRINI, J. L. et al. *Álgebra Linear*. 3ª ed. São Paulo: Harper & Row do Brasil, 1980.

# 1. Definição de Independência Linear

Considere um conjunto de vetores $S = \{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n\}$. 

A equação homogênea:
$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \ldots + c_n\mathbf{v}_n = \mathbf{0}$$

Sempre possui pelo menos uma solução trivial: $c_1 = c_2 = \ldots = c_n = 0$ (BOLDRINI et al., 1980).

- **Linearmente Independente (LI):** Se a equação acima possuir **APENAS** a solução trivial, o conjunto é LI. Nenhum vetor pode ser escrito como combinação dos outros.
- **Linearmente Dependente (LD):** Se existem outros escalares diferentes de zero que satisfaçam a equação, o conjunto é LD. Existe pelo menos um vetor que é redundante, ou seja, é apenas combinação linear dos demais (BOLDRINI et al., 1980).


# 2. Identificação via Escalonamento (RREF)

A maneira mais rigorosa e eficiente de identificar a independência linear de um conjunto de vetores é através da montagem de uma matriz, onde cada vetor ocupa uma coluna, seguida pela aplicação do **Escalonamento Reduzido (RREF)** (BOLDRINI et al., 1980).

- As colunas que contêm **pivôs** após o escalonamento indicam os vetores que compõem o conjunto **LI**.
- As colunas sem pivôs representam vetores que podem ser expressos como combinação linear dos vetores pivô anteriores, sendo, portanto, redundantes (**LD**).

Abaixo, apresentamos a implementação computacional utilizando a biblioteca SymPy.


```python
import sympy as sp

# Conjunto S = {v1, v2, v3, v4}
v1 = [1, 2, 0]
v2 = [2, 4, 0]  # Claramente v2 = 2*v1 (Logo v2 é LD)
v3 = [0, 0, 1]
v4 = [1, 2, 1]  # v4 = v1 + v3 (Logo v4 é LD)

A = sp.Matrix([v1, v2, v3, v4]).transpose()

display(A)

# Computando a Matriz em forma Reduzida por Linhas (RREF)
A_rref, pivotes = A.rref()

display(A_rref)
print(f"\nAs colunas pivô são (índices): {pivotes}")

print("\nConclusão:")
print("Apenas os vetores correspondentes às colunas com pivô (0 e 2) são Linearmente Independentes.")
print("Ou seja, os vetores v1 e v3!")
```

# 3. O Teste do Determinante

Se possuirmos um número de vetores igual à dimensão do espaço (por exemplo, 3 vetores no $\mathbb{R}^3$), a matriz $A$ será quadrada $3 \times 3$ (BOLDRINI et al., 1980).

- Se $\det(A) \neq 0 \implies$ Os vetores são **LI**.
- Se $\det(A) = 0 \implies$ Os vetores são **LD**.


```python
import numpy as np

# A matriz quadrada M
M = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])

det_M = np.round(np.linalg.det(M), 2)
display(sp.Matrix(M))
print(f"\nDeterminante de M = {det_M}")

if det_M != 0:
    print("Como det != 0, as colunas (ou linhas) de M são Linearmente Independentes!")
else:
    print("Como det == 0, as colunas (ou linhas) de M são Linearmente Dependentes!")
```
