## Introdução

Dentro de um espaço vetorial enorme (como o $\mathbb{R}^3$), podemos frequentemente encontrar "mundos menores" que também se comportam perfeitamente como espaços vetoriais. Esses mundos independentes são chamados de **Subespaços Vetoriais**.

Ao final deste tutorial, você será capaz de:
1. Definir formalmente um subespaço vetorial.
2. Identificar se um dado conjunto é ou não um subespaço.
3. Compreender a importância crucial de o subespaço **passar pela origem**.
4. Visualizar diferenças entre translações e subespaços com Python.


## 1. Definição Formal e Condições

Um subconjunto não vazio $W$ de um espaço vetorial $V$ é um **subespaço vetorial** de $V$ se o próprio $W$ for um espaço vetorial sob as mesmas regras.

Para testar, usamos o **Teste do Subespaço** (3 regras):
1. **O vetor nulo está em $W$:** $\mathbf{0} \in W$.
2. **Fechamento sob adição:** Se $\mathbf{u}$ e $\mathbf{v}$ estão em $W$, então $\mathbf{u} + \mathbf{v}$ estão em $W$.
3. **Fechamento sob multiplicação por escalar:** Se $\mathbf{u}$ está em $W$ e $c \in \mathbb{R}$, então $c\mathbf{u}$ está em $W$.

### Exemplos no $\mathbb{R}^3$
- Planos e retas infinitas **passando pela origem**.
- Qualquer plano ou reta transladado do zero NÃO é subespaço pois falta a origem.


## 2. Visualização: Subespaço vs translação affine

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

sns.set_theme(style="white", context="talk")

x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

Z_sub = X - Y        # Plano Z = X - Y (Passa na origem)
Z_nao_sub = X - Y + 5# Plano Z = X - Y + 5 (Transladado)

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z_sub, alpha=0.5, color='royalblue')
ax1.scatter([0], [0], [0], color='red', s=100, label='Origem (0,0,0)', zorder=5)
ax1.set_title("Subespaço: Passa pela Origem", pad=20)
ax1.legend()

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z_nao_sub, alpha=0.5, color='orange')
ax2.scatter([0], [0], [0], color='red', s=100, label='Origem (0,0,0)', zorder=5)
ax2.set_title("Transladado: NÃO é Subespaço", pad=20)
ax2.legend()

plt.tight_layout()
plt.show()
```

## 3. Prática: Núcleo matricial no SymPy
Soluções de $A\mathbf{x} = \mathbf{0}$ formam um subespaço no núcleo.

```python
import sympy as sp

A = sp.Matrix([
    [1, -2, 3],
    [2, -4, 6],
    [-1, 2, -3]
])
sp.pprint(A)

# A função nullspace retorna uma base do núcleo.
null_space = A.nullspace()

print("\nNúcleo de A (espaço de vetores x onde Ax = 0):")
for vec in null_space:
    sp.pprint(vec)
```
