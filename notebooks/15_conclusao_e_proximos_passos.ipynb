## 3. Tópicos Avançados e Próximos Passos

Sistemas lineares são a base para tópicos muito mais profundos:

### 3.1 Mínimos Quadrados (Ordinary Least Squares - OLS)
Quando temos mais equações que incógnitas (sistemas sobredeterminados) e os dados possuem ruído, geralmente não há solução exata ($Ax \neq b$). O método de **Mínimos Quadrados** busca o vetor $x$ que minimiza o erro $\|Ax - b\|^2$. É a base da regressão linear em Machine Learning.

### 3.2 Decomposições Matriciais (LU, QR, Cholesky)
Em computação de alto desempenho, raramente resolvemos $Ax=b$ via RREF. Usamos decomposições para "fatorar" a matriz $A$ em produtos de matrizes mais simples (triangulares ou ortogonais), otimizando a memória e o tempo de CPU.

### 3.3 Decomposição em Valores Singulares (SVD)
Considerada a "joia da coroa" da Álgebra Linear, a SVD permite decompor qualquer matriz, revelando suas propriedades fundamentais. É usada para compressão de imagens (PCA) e sistemas de recomendação.

## 4. Referências e Leituras Recomendadas

Para continuar seus estudos, recomendamos as seguintes fontes consagradas:

1.  **Gilbert Strang - "Linear Algebra and Its Applications":** O texto mais famoso da área, com foco em intuição e aplicações. [MIT OpenCourseWare](https://ocw.mit.edu/courses/mathematics/18-06-linear-algebra-spring-2010/).
2.  **Howard Anton & Chris Rorres - "Álgebra Linear com Aplicações":** Excelente para o rigor matemático balanceado com exemplos práticos.
3.  **Stephen Boyd - "Introduction to Applied Linear Algebra":** Foco total em aplicações computacionais, engenharia e otimização. Disponível online: [VMLS](http://vmls-book.stanford.edu/).
4.  **Documentação Oficial NumPy/SciPy:** Essencial para implementação profissional.
