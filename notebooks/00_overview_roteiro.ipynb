## 1. Introdução e Motivação

Sistemas lineares são o coração da Álgebra Linear e uma das ferramentas mais poderosas da matemática aplicada. Eles aparecem em praticamente todas as áreas do conhecimento, desde a engenharia e física até a economia, biologia e ciência de dados.

### Por que estudar Sistemas Lineares?
Resolver um sistema linear significa encontrar valores para variáveis que satisfaçam múltiplas condições (equações) ao mesmo tempo. No mundo real, isso se traduz em:
- **Engenharia de Tráfego:** Equilibrar o fluxo de veículos em uma malha urbana.
- **Circuitos Elétricos:** Calcular correntes e tensões em redes complexas.
- **Ciência de Dados:** Ajustar modelos de regressão ou realizar processamento de imagens.
- **Economia:** Analisar o equilíbrio de oferta e demanda em mercados interconectados.

## 4. Exemplo Motivador: A Simplicidade do Python

Para sentir o gostinho do que faremos, veja como é simples resolver um sistema linear $2 \times 2$ usando Python. 

Imagine o seguinte problema: "A soma de dois números é 10 e a diferença entre eles é 2. Quais são os números?"
O sistema é:
$$
\begin{cases}
x + y = 10 \\
x - y = 2
\end{cases}
$$


```
import numpy as np
import matplotlib.pyplot as plt

# 1. Definindo as matrizes do sistema Ax = b
A = np.array([[1, 1], 
              [1, -1]])
b = np.array([10, 2])

# 2. Resolvendo o sistema
x_sol = np.linalg.solve(A, b)
print(f"Solução encontrada: x = {x_sol[0]}, y = {x_sol[1]}")

# 3. Visualização rápida
x_vals = np.linspace(0, 10, 100)
y1 = 10 - x_vals
y2 = x_vals - 2

plt.figure(figsize=(6, 4))
plt.plot(x_vals, y1, label='x + y = 10')
plt.plot(x_vals, y2, label='x - y = 2')
plt.scatter(x_sol[0], x_sol[1], color='red', zorder=5, label=f'Interseção {x_sol}')
plt.axhline(0, color='black', lw=1); plt.axvline(0, color='black', lw=1)
plt.legend(); plt.grid(True, linestyle='--')
plt.title("Visualização do Sistema 2x2")
plt.show()
```
