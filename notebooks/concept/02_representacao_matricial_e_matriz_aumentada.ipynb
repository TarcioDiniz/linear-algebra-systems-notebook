## Representação Matricial de Sistemas Lineares

Neste notebook veremos como escrever sistemas lineares
na forma matricial e introduzir o conceito de matriz aumentada.

### Matrizes

Uma **matriz** é uma tabela retangular de números
organizados em linhas e colunas.

Exemplo:

$$
A =\begin{bmatrix}
2 & 1  \\
1 & -1 \\
\end{bmatrix}
$$


### Forma Matricial

Considere o sistema
$$
\begin{cases}
2x + y = 5  \\
x − y = 1
\end{cases}
$$


Podemos escrever:
$$
Ax = b
$$

### Matriz Aumentada

A matriz aumentada é obtida anexando o vetor b à matriz A.

$$
\begin{array}{|c|c|}
A & b
\end{array}
$$

Exemplo
$$
\begin{array}{|cc|c|}
2 & 1 & 5 \\
1 & -1 & 1
\end{array}
$$


```python
import numpy as np

A = np.array([
[2,1],
[1,-1]
])

b = np.array([5,1])

aug = np.column_stack((A,b))

print(aug)
```

### Posto da Matriz

O **posto** de uma matriz corresponde ao número máximo
de linhas ou colunas linearmente independentes.

```python
np.linalg.matrix_rank(A)
```

### Classificação do Sistema

Se
$$
posto(A) = posto(A|b) = número de variáveis
\\
→ SPD
$$
Se

$$
posto(A) = posto(A|b) < número de variáveis
\\
→ SPI
$$

Se
$$
posto(A) ≠ posto(A|b)
\\
→ SI
$$

```python
def classificar(A,b):

    aug = np.column_stack((A,b))

    rA = np.linalg.matrix_rank(A)
    rAug = np.linalg.matrix_rank(aug)

    n = A.shape[1]

    if rA == rAug == n:
        return "SPD"

    elif rA == rAug:
        return "SPI"

    else:
        return "SI"
    
print(classificar(A,b))
```
