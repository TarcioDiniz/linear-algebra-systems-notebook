## Introdução

Os sistemas lineares aparecem em diversas áreas da matemática,
engenharia, computação e ciência de dados.

De forma geral, um sistema linear é um conjunto de equações
que devem ser satisfeitas simultaneamente.

Problemas como:

- balanceamento químico
- análise de circuitos elétricos
- fluxo de tráfego
- ajuste de modelos matemáticos

podem ser representados através de sistemas lineares.


### Equação Linear

Uma **equação linear** em n variáveis é uma equação da forma:

`a₁x₁ + a₂x₂ + ... + aₙxₙ = b`

onde:

- a₁, a₂, ..., aₙ são números reais (coeficientes)
- x₁, x₂, ..., xₙ são as variáveis
- b é o termo independente

Exemplo:

$$
2x + y = 5
$$

### Sistema Linear

Um **sistema linear** é um conjunto de equações lineares
envolvendo as mesmas variáveis.

Exemplo:
$$
\begin{cases}
2x + y = 5 \\
x − y = 1
\end{cases}
$$

A solução do sistema é o conjunto de valores das variáveis
que satisfazem simultaneamente todas as equações.

### Conjunto Solução

O **conjunto solução** de um sistema linear é o conjunto de todos
os valores das variáveis que tornam verdadeiras todas as equações
do sistema.

Exemplo:

Sistema:

$$
\begin{cases}
2x + y = 5 \\
x − y = 1
\end{cases}
$$


Solução:

$$
x = 2  \\
y = 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

y1 = 5 - 2*x
y2 = x - 1

plt.plot(x,y1,label="2x + y = 5")
plt.plot(x,y2,label="x - y = 1")

plt.legend()
plt.grid()
plt.show()
```

### Classificação dos Sistemas Lineares

Os sistemas lineares podem ser classificados em três tipos:

#### Sistema Possível Determinado (SPD)

- Possui **uma única solução**.

### Sistema Possível Indeterminado (SPI)

- Possui **infinitas soluções**.

### Sistema Impossível (SI)

- **Não possui solução.**

### Interpretação Geométrica

Em sistemas com duas variáveis, cada equação representa
uma **reta no plano cartesiano**.

Existem três possibilidades:

1. As retas se intersectam em um ponto  
→ Sistema Possível Determinado

2. As retas coincidem  
→ Sistema Possível Indeterminado

3.  As retas são paralelas  
→ Sistema Impossível

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

y1 = 5 - 2*x
y2 = x - 1

plt.figure()

plt.plot(x, y1, label="2x + y = 5")
plt.plot(x, y2, label="x - y = 1")

plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.grid()

plt.title("Interpretação Geométrica de um Sistema Linear")

plt.show()
```
