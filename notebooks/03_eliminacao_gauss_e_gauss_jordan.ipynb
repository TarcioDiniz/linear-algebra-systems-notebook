# 1. Operações Elementares de Linha

Para resolver um sistema linear, aplicamos transformações que alteram as equações, mas preservam o **conjunto solução**. Dois sistemas são ditos **equivalentes** se possuem exatamente as mesmas soluções (BOLDRINI et al., 1980).

As três operações elementares de linha são:

1.  **Troca de linhas ($L_i \leftrightarrow L_j$):** Permutar a posição de duas equações.
2.  **Multiplicação por escalar não nulo ($L_i \leftarrow k L_i, k \neq 0$):** Multiplicar todos os termos de uma linha por uma constante.
3.  **Combinação linear de linhas ($L_i \leftarrow L_i + k L_j$):** Somar a uma linha o múltiplo de outra.

---

# 2. Forma Escalonada (REF)

Uma matriz está na **forma escalonada** quando (BOLDRINI et al., 1980):
-   Linhas nulas (se houver) estão na parte inferior.
-   O primeiro elemento não nulo de cada linha (chamado de **pivô**) está à direita do pivô da linha superior.

### Exemplo de Matriz na Forma Escalonada:
$$
\begin{pmatrix}
\mathbf{2} & 3 & 1 \\
0 & \mathbf{5} & -2 \\
0 & 0 & 0
\end{pmatrix}
$$
Os pivôs estão em negrito. Eles indicam as **variáveis líderes**. As colunas sem pivô correspondem às **variáveis livres**.

---

# 3. Forma Escalonada Reduzida (RREF)

Uma matriz está na **forma escalonada reduzida** se, além de estar na forma escalonada, satisfaz (BOLDRINI et al., 1980):
-   Todos os pivôs são iguais a **1**.
-   Em cada coluna que contém um pivô, todos os outros elementos são **0**.

### Exemplo de Matriz na Forma Escalonada Reduzida:
$$
\begin{pmatrix}
\mathbf{1} & 0 & 5 \\
0 & \mathbf{1} & -2 \\
0 & 0 & 0
\end{pmatrix}
$$
A RREF de uma matriz é **única**.

---

# 4. Eliminação de Gauss

O algoritmo de **Eliminação de Gauss** transforma a matriz aumentada do sistema em uma matriz na **forma escalonada (REF)** através de operações elementares (BOLDRINI et al., 1980).

### Algoritmo Passo a Passo:
1.  Identificar a coluna mais à esquerda que não seja toda nula.
2.  Escolher um pivô não nulo nessa coluna (trocando linhas se necessário).
3.  Zerar todos os elementos abaixo do pivô usando a operação de combinação linear.
4.  Repetir o processo para a submatriz abaixo e à direita do pivô atual.
5.  Finalizar com a **substituição regressiva** para encontrar os valores das incógnitas.

---

## 5. Eliminação de Gauss-Jordan

A **Eliminação de Gauss-Jordan** é uma extensão que continua o processo de eliminação acima dos pivôs e normaliza-os para 1, resultando diretamente na **forma escalonada reduzida (RREF)**.

-   **Vantagem:** A solução (ou a estrutura da solução) pode ser lida diretamente da matriz final.
-   **Custo:** Exige mais operações aritméticas que a Eliminação de Gauss clássica, mas é conceitualmente mais poderosa para análise estrutural.

---

## 6. Classificação de Sistemas via Escalonamento

Após o escalonamento da matriz aumentada $[A|b]$, podemos classificar o sistema:

-   **SI (Sistema Impossível):** Se houver uma linha da forma $[0 \ 0 \ \dots \ 0 \ | \ k]$ com $k \neq 0$ (ex: $0 = 5$).
-   **SPD (Sistema Possível Determinado):** Se não houver a linha de inconsistência e cada coluna da matriz de coeficientes tiver um pivô (zero variáveis livres).
-   **SPI (Sistema Possível Indeterminado):** Se não houver a linha de inconsistência e houver pelo menos uma coluna sem pivô (pelo menos uma variável livre).

---

## 7. Aspectos Computacionais: Pivotamento Parcial

Em computação de ponto flutuante, escolher um pivô muito pequeno pode gerar erros de arredondamento catastróficos.
O **pivotamento parcial** consiste em trocar a linha atual pela linha que possui o elemento de maior valor absoluto na coluna do pivô. Isso aumenta a estabilidade numérica do algoritmo.

---

## 8. Demonstração Computacional com SymPy

O SymPy permite realizar o escalonamento de forma simbólica, o que é ideal para fins didáticos, pois evita erros de arredondamento e mantém frações.

### Exemplo: Resolvendo um Sistema 3x3

```python
import sympy as sp

# Definindo as variáveis
x, y, z = sp.symbols('x y z')

# Sistema:
#  x + 2y +  z = 2
# 2x + 6y +  z = 7
#  x +  y + 4z = 3

A = sp.Matrix([
    [1, 2, 1],
    [2, 6, 1],
    [1, 1, 4]
])

b = sp.Matrix([2, 7, 3])

# Matriz aumentada
aug = A.row_join(b)
display(aug)

# Obtendo a RREF e os índices dos pivôs
rref_matrix, pivot_indices = aug.rref()

display(rref_matrix)
print(f"\nÍndices das colunas com pivôs: {pivot_indices}")

# Resolvendo o sistema
solucao = sp.solve_linear_system(aug, x, y, z)
display(solucao)
```

## 9. Visualização do Processo Passo a Passo

Abaixo, implementamos uma função simples para visualizar as operações de linha.

```python
def visualizar_escalonamento():
    M = sp.Matrix([
        [2, 1, -1, 8],
        [-3, -1, 2, -11],
        [-2, 1, 2, -3]
    ])
    
    display(M)
    
    # Passo 1: L2 = L2 + (3/2)*L1
    M[1, :] = M[1, :] + sp.Rational(3, 2) * M[0, :]
    display(M)
    
    # Passo 2: L3 = L3 + L1
    M[2, :] = M[2, :] + M[0, :]
    display(M)
    
    # Passo 3: L3 = L3 - 4*L2
    M[2, :] = M[2, :] - 4 * M[1, :]
    display(M)
    
    # Agora a matriz está em REF (quase, falta normalizar se quisermos RREF)
    # L3 = L3 * (-1/5)
    M[2, :] = M[2, :] * sp.Rational(-1, 5)
    display(M)

visualizar_escalonamento()
```

## 10. Posto e Variáveis Livres

A forma escalonada nos revela propriedades fundamentais do espaço das soluções:

1.  **Posto da Matriz ($rank$):** É o número de linhas não nulas na forma escalonada (ou o número de pivôs).
2.  **Variáveis Líderes:** Associadas às colunas com pivô.
3.  **Variáveis Livres:** Associadas às colunas sem pivô. Representam os "graus de liberdade" do sistema.
    -   Se $rank(A) = rank([A|b]) = \text{número de variáveis}$, o sistema é **SPD**.
    -   Se $rank(A) = rank([A|b]) < \text{número de variáveis}$, o sistema é **SPI**.
    -   Se $rank(A) < rank([A|b])$, o sistema é **SI**.
