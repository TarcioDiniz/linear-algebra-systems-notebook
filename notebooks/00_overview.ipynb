## 2. Sistema Linear

Um sistema linear é um conjunto de equações da forma

a11 x1 + a12 x2 + ... + a1n xn = b1  
a21 x1 + a22 x2 + ... + a2n xn = b2  
...  
am1 x1 + am2 x2 + ... + amn xn = bm

onde:

- x1, x2, ..., xn são incógnitas
- aij são coeficientes
- bi são termos independentes

### Exemplo

2x + y = 5  
x − y = 1

## 3. Representação Matricial

Um sistema linear pode ser representado de forma compacta por

A x = b

onde

A = matriz de coeficientes  
x = vetor de incógnitas  
b = vetor de termos independentes

### Forma matricial

A = [aij]

x = (x1, x2, ..., xn)

b = (b1, b2, ..., bm)

## 4. Matriz Aumentada

Para estudar um sistema linear utilizamos frequentemente a matriz aumentada

[A | b]

Essa representação permite aplicar operações elementares de linha para transformar o sistema em uma forma mais simples.

## 5. Operações Elementares de Linha

As seguintes operações preservam o conjunto solução:

1. Trocar duas linhas
2. Multiplicar uma linha por escalar não nulo
3. Somar múltiplo de uma linha a outra

Essas operações são utilizadas no método de eliminação de Gauss.

## 6. Eliminação de Gauss

O método de eliminação de Gauss transforma a matriz aumentada em forma escalonada.

O objetivo é obter um sistema equivalente mais simples para determinar as soluções.

Passos:

1. escolher pivô
2. eliminar coeficientes abaixo
3. repetir para as linhas seguintes

## 7. Classificação dos Sistemas Lineares

Um sistema pode ser classificado em:

### Sistema Possível Determinado (SPD)

possui solução única.

### Sistema Possível Indeterminado (SPI)

possui infinitas soluções.

### Sistema Impossível (SI)

não possui solução.

## 8. Interpretação Geométrica

Em duas variáveis, cada equação representa uma reta no plano.

Possibilidades:

- uma interseção → solução única
- retas paralelas → nenhuma solução
- retas coincidentes → infinitas soluções

## 9. Sistemas Homogêneos

Um sistema homogêneo tem a forma

A x = 0

Esse sistema sempre possui a solução trivial

x = 0

Quando existem soluções não triviais, elas formam um subespaço vetorial chamado núcleo da matriz.

## 10. Dependência Linear

Um conjunto de vetores é linearmente dependente se existe uma combinação linear não trivial que resulta no vetor nulo.

Isso está diretamente relacionado à existência de soluções não triviais para

A x = 0

## 11. Sistemas Lineares como Transformações Lineares

Toda matriz A define uma transformação linear

T(x) = A x

Resolver o sistema

A x = b

equivale a encontrar um vetor x cuja imagem pela transformação T seja o vetor b.

## 12. Aplicação: Fluxo de Tráfego

## 13. Aplicação: Interpolação Polinomial

Dado um conjunto de pontos

(x1,y1), (x2,y2), ..., (xn,yn)

podemos determinar um polinômio que passa por esses pontos resolvendo um sistema linear.

Os coeficientes do polinômio são as incógnitas do sistema.
