## Introdução

Muitas vezes, ao colecionar vetores para formar um Span, adicionamos vetores "inúteis" que não trazem nenhuma nova direção para a coleção. Eles poderiam ser formados pelos vetores que já tínhamos! Quando isso acontece, o conjunto é **Linearmente Dependente (LD)**. Se todos os vetores forem vitais e trouxerem uma direção única, o conjunto é **Linearmente Independente (LI)**.

Ao final deste tutorial, você será capaz de:
1. Definir formalmente os conceitos de LI e LD.
2. Identificar vetores redundantes numa matriz.
3. Usar o escalonamento matrizial no Python (RREF) para extrair vetores linearmente independentes.
4. Conectar estes conceitos com determinantes.


## 1. A Definição Formal de LI e LD

Considere um conjunto de vetores $S = \{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n\}$. 

A equação homogênea:
$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \ldots + c_n\mathbf{v}_n = \mathbf{0}$$

Sempre possui pelo menos uma solução trivial: $c_1 = c_2 = \ldots = c_n = 0$.

- **Linearmente Independente (LI):** Se a equação acima possuir **APENAS** a solução trivial, o conjunto é LI. Nenhum vetor pode ser escrito como combinação dos outros.
- **Linearmente Dependente (LD):** Se existem outros escalares diferentes de zero que satisfaçam a equação, o conjunto é LD. Existe pelo menos um vetor que é "peso morto", ou seja, é apenas combinação linear dos demais.


## 2. A Mágica do RREF (Row-Reduced Echelon Form)

A maneira mais prática de descobrir se um conjunto de vetores é LI ou LD (e sobretudo *quais* são os LDs) é colocá-los nas colunas de uma matriz e aplicar o **Escalonamento Reduzido (RREF)**.

- As colunas que possuírem **pivôs** (linhas com 1 na diagonal e zeros acima e abaixo) indicam os vetores **LI** originais.
- As colunas sem pivô são as que dependem linearmente das anteriores (os vetores **LD**).

Vamos fazer isso com SymPy.


```python
import sympy as sp

# Conjunto S = {v1, v2, v3, v4}
v1 = [1, 2, 0]
v2 = [2, 4, 0]  # Claramente v2 = 2*v1 (Logo v2 é LD)
v3 = [0, 0, 1]
v4 = [1, 2, 1]  # v4 = v1 + v3 (Logo v4 é LD)

A = sp.Matrix([v1, v2, v3, v4]).transpose()

print("Matriz A (vetores nas colunas):")
sp.pprint(A)

# Computando a Matriz em forma Reduzida por Linhas (RREF)
A_rref, pivotes = A.rref()

print("\nMatriz A na forma RREF:")
sp.pprint(A_rref)
print(f"\nAs colunas pivô são (índices): {pivotes}")

print("\nConclusão:")
print("Apenas os vetores correspondentes às colunas com pivô (0 e 2) são Linearmente Independentes.")
print("Ou seja, os vetores v1 e v3!")
```

## 3. O Teste Rápido do Determinante
Se possuirmos um número de vetores igual à dimensão do espaço (por exemplo, 3 vetores no $\mathbb{R}^3$), a matriz A será quadrada $3 \times 3$.

- Se $\det(A) \neq 0 \implies$ Os vetores são **LI**.
- Se $\det(A) = 0 \implies$ Os vetores são **LD**.


```python
import numpy as np

# A matriz quadrada M
M = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])

det_M = np.round(np.linalg.det(M), 2)
print("Matriz M:")
print(M)
print(f"\nDeterminante de M = {det_M}")

if det_M != 0:
    print("Como det != 0, as colunas (ou linhas) de M são Linearmente Independentes!")
else:
    print("Como det == 0, as colunas (ou linhas) de M são Linearmente Dependentes!")
```
