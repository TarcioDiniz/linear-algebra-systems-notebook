## Introdução

Neste tutorial, consolidaremos os conceitos de **Base** e **Dimensão**. A base de um espaço vetorial representa o conjunto mínimo de vetores linearmente independentes capaz de gerar todo o espaço, servindo como o alicerce para a definição de coordenadas e transformações.

Ao final deste tutorial, o estudante será capaz de:
1. Definir formalmente o conceito de Base.
2. Determinar a base e a dimensão de um subespaço vetorial.
3. Calcular as coordenadas de um vetor em relação a diferentes bases.
4. Implementar matrizes de mudança de base utilizando Python.

**Referência principal:** BOLDRINI, J. L. et al. *Álgebra Linear*. 3ª ed. São Paulo: Harper & Row do Brasil, 1980.

# 1. Definição de Base

Um conjunto $B = \{\mathbf{v}_1, \ldots, \mathbf{v}_n\}$ é chamado de **Base** para um espaço vetorial $V$ (ou para um subespaço) se cumprir DUAS condições simultaneamente (BOLDRINI et al., 1980):

1. **O conjunto $B$ é Linearmente Independente (LI)**. (Não possui redundâncias).
2. **O conjunto $B$ gera $V$**. (Seu "Span" consegue alcançar todos os elementos do espaço $V$).

### O que é Dimensão?
Todas as possíveis bases para um mesmo espaço vetorial $V$ possuem **exatamente o mesmo número de vetores**. 
Esse número de vetores que compõem uma base é justamente a **Dimensão** do espaço (denotada por $\dim V$) (BOLDRINI et al., 1980). 

- A dimensão do $\mathbb{R}^n$ é $n$.
- A Base Canônica do $\mathbb{R}^3$ é $B = \{ (1,0,0), (0,1,0), (0,0,1) \}$.


# 2. Base e Dimensão via Python

Suponha que somos apresentados a um subespaço de $\mathbb{R}^4$ gerado por 5 vetores:
$\langle \mathbf{v}_1, \mathbf{v}_2, \mathbf{v}_3, \mathbf{v}_4, \mathbf{v}_5 \rangle$

Como filtramos esses 5 vetores para extrair apenas uma base verdadeira? E qual será a dimensão deste subespaço? Basta montar uma matriz e realizar a eliminação de Gauss (RREF) para extrair a base.


```python
import sympy as sp

# 5 vetores no R^4
v1 = [1, 2, 0, 1]
v2 = [2, 4, 0, 2] # múltiplo de v1
v3 = [-1, 3, 2, 0]
v4 = [0, 5, 2, 1] # é v1 + v3
v5 = [1, 0, 1, 0]

# Colocamos os vetores nas COLUNAS da nossa Matriz A
A = sp.Matrix([v1, v2, v3, v4, v5]).transpose()

display(A)

# Achando os pivôs com RREF
_, pivotes = A.rref()

print(f"\nOs índices das colunas pivô extraídas são: {pivotes}")

# A dimensão é a quantidade de vetores na base!
dimensao = len(pivotes)

print(f"\nA dimensão do subespaço é = {dimensao}")
print("Uma Base válida é formada pelas colunas da matriz original referentes aos pivôs:")

base = [A.col(p) for p in pivotes]
for i, vec in enumerate(base):
    display(vec)
```

## 3. Coordenadas em Novas Bases

Qualquer vetor $\mathbf{u}$ pode ser escrito como combinações dos vetores da base $B$. Os escalares necessários para formá-lo são as famosas **coordenadas** do vetor em relação a essa base.

Se a Base $B$ forma uma matriz $P$ (juntando os vetores nas colunas), as coordenadas do vetor $\mathbf{u}$ na base $B$ (que chamamos de $\mathbf{x}_B$) são encontradas resolvendo:

$$P \mathbf{x}_B = \mathbf{u} \implies \mathbf{x}_B = P^{-1} \mathbf{u}$$


```python
import numpy as np

import sympy as sp

# A nova Base B (2x2) no R^2
# Vetores (1,1) e (-1,1)
P = np.array([
    [1, -1],
    [1,  1]
])

# Vetor dado nas coordenadas tradicionais (Base Canônica)
u = np.array([4, 2])

# Para encontrar as coordenadas na Base P, multiplicamos pela inversa de P
u_na_base_P = np.linalg.inv(P).dot(u)

display(sp.Matrix(P))

display(sp.Matrix(u))
print(f"Coordenadas do vetor u em relação à Base P:")
display(sp.Matrix(u_na_base_P))
```
